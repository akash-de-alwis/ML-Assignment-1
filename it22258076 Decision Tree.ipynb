```python
# ── Imports ──────────────────────────────────────────────────────────────────
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import (accuracy_score, confusion_matrix,
                             classification_report, ConfusionMatrixDisplay)

# Make plots look nice
sns.set_theme(style="whitegrid")
%matplotlib inline
```

```python
# ── Load Dataset ─────────────────────────────────────────────────────────────
df = pd.read_csv("Titanic_Dataset.csv")
print("Shape:", df.shape)
df.head()
```

```python
# ── Explore the Data ─────────────────────────────────────────────────────────
print("=== Column Info ===")
df.info()
print()
print("=== Missing Values ===")
print(df.isnull().sum())
print()
print("=== Survival Count ===")
print(df["survived"].value_counts())
```

```python
# ── Clean the Data ───────────────────────────────────────────────────────────

# 1. Drop columns that are not useful or would cause data leakage
df = df.drop(columns=["name","ticket","cabin","boat","body","home.dest"])

# 2. Fill missing Age with the median age
df["age"] = df["age"].fillna(df["age"].median())

# 3. Fill missing Fare with the median fare
df["fare"] = df["fare"].fillna(df["fare"].median())

# 4. Fill missing Embarked with the most common port
df["embarked"] = df["embarked"].fillna(df["embarked"].mode()[0])

# 5. Convert Sex to numbers: male=0, female=1
df["sex"] = df["sex"].map({"male":0 , "female":1})

# 6. Convert Embarked to numbers: S=0, C=1, Q=2
df["embarked"] = df["embarked"].map({"S":0 , "C":1 , "Q":2})

print("Cleaned dataset shape:", df.shape)
print("Any missing values left?", df.isnull().sum().any())
df.head()
```
